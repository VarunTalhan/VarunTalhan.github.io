```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def square(t, duty=0.5, amp=1):
    return amp*signal.square(2*np.pi*t, duty)

def triangular(t, width=0.5, amp=1):
    return amp*signal.sawtooth(2*np.pi*t, width)

def sine(t, amp=1):
    return amp*np.sin(2*np.pi*t)
```

```python
t = np.linspace(0, 5, 1000)
y_square = square(t)
y_trig = triangular(t)
y_sin = sine(t)

fig, ax = plt.subplots()
ax.plot(t, y_square, label='sqaure')
ax.plot(t, y_trig, label='triangular')
ax.plot(t, y_sin, label='sine')
ax.set_xlabel('t')
ax.set_ylabel('y')
ax.legend()
```

```python
def add_noise(y, std):
    y_noisy = y+np.random.normal(0, std, len(y))
    return y_noisy

y_sin_noise = add_noise(y_sin, 0.1)

fig, ax = plt.subplots()
ax.plot(t, y_sin, label='no noise')
ax.plot(t, y_sin_noise, zorder=-1, label='with noise')
ax.set_xlabel('t')
ax.set_xlabel('y')
ax.legend()
```

```python
pd.DataFrame({
    't' : t,
    'y' : y_square
}).to_csv('square_wave.csv', index=False)
pd.DataFrame({
    't' : t,
    'y' : y_trig
}).to_csv('triangular_wave.csv', index=False)
pd.DataFrame({
    't' : t,
    'y' : y_sin
}).to_csv('sine_wave.csv', index=False)
pd.DataFrame({
    't' : t,
    'y' : y_sin_noise
}).to_csv('noisy_sine_wave.csv', index=False)
```
